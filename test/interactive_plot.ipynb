```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
f,a = plt.subplots()
x = np.linspace(1,10,100)
y = np.sin(x)
a.plot(x,y)
pos = []
def onclick(event):
    print('Processing click...')
    pos = [event.xdata,event.ydata]
    a.set_title('%s' % pos)
f.canvas.mpl_connect('button_press_event', onclick)
# f.show()
```
